```python
prefixos = ['PlacaPedra', 'Placa', 'Grama', 'Grama45', 'Pedra90', 'Pedra45']
colunas_filtradas = [col for col in merged_df.columns if any(col.startswith(p) for p in prefixos)]

# Calcular a mdia e o desvio padro para as colunas filtradas
df_stats = merged_df[['Wavelength']].copy()

for prefixo in prefixos:
  cols_com_prefixo = [col for col in colunas_filtradas if col.startswith(prefixo)]
  if cols_com_prefixo:
    df_stats[f'{prefixo}_mean'] = merged_df[cols_com_prefixo].mean(axis=1)
    df_stats[f'{prefixo}_std'] = merged_df[cols_com_prefixo].std(axis=1)

# Claculando a reflectncia
df_reflect = pd.DataFrame()
df_reflect['Wavelength'] = df_stats['Wavelength']
df_reflect['grama90'] = df_stats['Grama_mean'] / df_stats['Placa_mean']
df_reflect['pedra90'] = df_stats['Pedra90_mean'] / df_stats['PlacaPedra_mean']
df_reflect['grama45'] = df_stats['Grama45_mean'] / df_stats['Placa_mean']
df_reflect['pedra45'] = df_stats['Pedra45_mean'] / df_stats['PlacaPedra_mean']

df_reflect.head()
```

```python
#buscando imagens sentinel para o perodo
def mask_s2_clouds(image):
  """Masks clouds in a Sentinel-2 image using the QA band.
      image (ee.Image): A Sentinel-2 image.
      ee.Image: A cloud-masked Sentinel-2 image.
  """
  qa = image.select('QA60')

  # Bits 10 and 11 are clouds and cirrus, respectively.
  cloud_bit_mask = 1 << 10
  cirrus_bit_mask = 1 << 11

  # Both flags should be set to zero, indicating clear conditions.
  mask = (
      qa.bitwiseAnd(cloud_bit_mask)
      .eq(0)
      .And(qa.bitwiseAnd(cirrus_bit_mask).eq(0)))
  return image.updateMask(mask).divide(10000)

#localizao do INPE
longitude = -45.858087
latitude = -23.209335
coordenadas = ee.Geometry.Point(longitude, latitude)
ROI = coordenadas.buffer(10000)

dataset = (
    ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
    .filterDate('2025-04-02', '2025-04-10')
    # Pre-filter to get less cloudy granules.
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))
    .map(mask_s2_clouds)
    .filterBounds(coordenadas))

visualization = {
    'min': 0.0,
    'max': 0.3,
    'bands': ['B4', 'B3', 'B2'],}

Map = geemap.Map(center=[latitude, longitude], zoom=10)
Map.addLayer(dataset.mean().clip(ROI), visualization, 'RGB')
Map.addLayer(coordenadas, {}, 'Localizao')
Map
```

```python
# Montando o ambiente ()
!pip install py6s -q
!apt install gfortran libgfortran5 -q
!wget https://anaconda.org/conda-forge/sixs/1.1.3/download/linux-64/sixs-1.1.3-h2cc385e_2.tar.bz2 -q
!tar -xvf sixs-1.1.3-h2cc385e_2.tar.bz2
!rm sixs-1.1.3-h2cc385e_2.tar.bz2
!chmod +x bin/sixs
!ln bin/sixs /usr/local/bin/sixs
!pip install rasterio -q

import rasterio
from Py6S import *
import numpy as np
import pandas as pd
import glob
from osgeo import gdal
gdal.UseExceptions()
import matplotlib.pyplot as plt
```

```python
# Variaveis de Entrada:
# Para acessar as variveis de entrada da sua rea, sigo o tutorial do link:
# https://github.com/Liaons/py6s-correcao-atmosferica/blob/main/parametros_atmosfericos-GEE.pdf

# Atmosfera + alvo:
AOD550 = 0.144 # Medida adquirida de satlite, MODIS
Altitude = 0.618 # ajuste para altitude do alvo, em km

# Imagem:
dia = 10
mes = 4
Geometria_solar_AZ = solar_a_angles
Geometria_solar_Zn = solar_z_angles
Geometria_sensor_AZ = mean_a_angle
Geometria_sensor_Zn = mean_z_angle

s = SixS()
# Definindo os parametros do 6SV:
s.atmos_profile = AtmosProfile.PredefinedType(AtmosProfile.Tropical) # definifir os valores de vapor d'agua e temperatura do ar esperados para a imagem.
# Carga optica do aerossol and Perfil/Modelo do aerossol:
s.aot550 = AOD550
s.aero_profile = AeroProfile.PredefinedType(AeroProfile.Continental) # relativo a distribuicao e composicao fisico-quimica do aerosol na atmosfera;
# Geometrias de iluminacao (solar) e visada (sensor):
s.geometry = Geometry.User()
s.geometry.day = dia
s.geometry.month = mes
s.geometry.solar_z = Geometria_solar_Zn # ngulo zenital solar da metadata
s.geometry.solar_a = Geometria_solar_AZ # azimute solar da metadata
s.geometry.view_z = Geometria_sensor_Zn # ngulo zenital de visada da metadata
s.geometry.view_a = Geometria_sensor_AZ # azimute de visada da metadata
# Altitude:
s.altitudes = Altitudes()
s.altitudes.set_sensor_satellite_level()  # nivel do sensor;
s.altitudes.set_target_custom_altitude(Altitude)  # nivel do alvo (em km).

# Configurar os comprimentos de ondas das bandas espectrais:
# Interpolacao da Funcao de Resposta Espectral:
def interpolar_bandas(rsr, band_number: int, range_w: list) -> dict:
    output = {}
    bands = s2_band
    for i in range(1, band_number + 1):
        filter_ = rsr.loc[rsr['id'] == i]
        w_min_ = np.min(filter_['wavelength'])
        w_max_ = np.max(filter_['wavelength'])
        wavelength_ = [np.around(w, 3) for w in np.arange(w_min_, w_max_, 0.001)]
        wavelength_int = [np.around(w, 3) for w in np.arange(range_w[0], range_w[1], 0.001)]
        wavelength_sub = list(set(wavelength_int) - set(wavelength_))
        df_ = pd.DataFrame({'wavelength': wavelength_sub, 'rsr': int(0), 'band': str(i)})
        df_ = pd.concat([filter_, df_]).sort_values(by=['wavelength'], ascending=True)
        d_select = df_['rsr'].to_list()
        wavelength_interp = [np.around(w, 4) for w in np.arange(range_w[0], range_w[1], 0.0025)]
        norm = np.interp(wavelength_interp, df_['wavelength'].to_list(), d_select).tolist()
        out_ = [norm, [i for i in wavelength_interp]]
        output[bands[i - 1]] = out_
    return output

url = "https://raw.githubusercontent.com/Liaons/py6s-correcao-atmosferica/main/rsr_S2B_MSI.txt"
rsr = pd.read_csv(url, sep='\t', skiprows=5, names=['wavelength', 'rsr', 'band', 'id'])

bandas = interpolar_bandas(rsr, 13, [0.411, 2.383])

# Modelagem atmosferica com o 6SV:
atm_parametros = {}

for i in s2_band:
    s.wavelength = Wavelength(np.min(bandas[i][1]), np.max(bandas[i][1]), bandas[i][0]) # min comprimento de onda, max comp.onda, rsr (2.5 nm).
    s.run()
    atm_parametros[i] = {'tg_OG_co': float(s.outputs.transmittance_co.total),
                         'tg_OG_c02': float(s.outputs.transmittance_co2.total),
                         'tg_OG_o2': float(s.outputs.transmittance_oxygen.total),
                         'tg_OG_no2': float(s.outputs.transmittance_no2.total),
                         'tg_OG_ch4': float(s.outputs.transmittance_ch4.total),
                         'Tg_O3': float(s.outputs.transmittance_ozone.total),
                         'Tg_H20': float(s.outputs.transmittance_water.total),
                         'T_upward': float(s.outputs.transmittance_total_scattering.upward),
                         'T_downward': float(s.outputs.transmittance_total_scattering.downward),
                         'p_atm': float(s.outputs.atmospheric_intrinsic_reflectance),
                         's': float(s.outputs.spherical_albedo.total)}
atm_parametros

# Transmitancias:
transm = []
#atm = []
for i in s2_band:
    transm.append(atm_parametros[i]['T_upward'] * atm_parametros[i]['T_downward'])
    #atm.append(atm_parametros[i]['p_atm'])

plt.plot(s2_band, transm, marker='o', linestyle='-')
#plt.plot(s2_band, atm, marker='o', linestyle='-')
```

```python
def abrir_imagem(path):
    return gdal.Open(path).ReadAsArray().astype(float)

def salvar_imagem(array, referencia, destino):
    ds_ref = gdal.Open(referencia)
    if ds_ref is None:
        print('Erro ao abrir referncia:', referencia)
        return
    driver = gdal.GetDriverByName('GTiff')
    out_path = os.path.join(destino, os.path.basename(referencia).replace('.tif','_corr.tif'))
    out_ds = driver.Create(out_path, ds_ref.RasterXSize, ds_ref.RasterYSize, 1, gdal.GDT_Float32)
    out_ds.SetGeoTransform(ds_ref.GetGeoTransform())
    out_ds.SetProjection(ds_ref.GetProjection())
    out_ds.GetRasterBand(1).WriteArray(array)
    out_ds.FlushCache()
    out_ds = None
    print('Salvo:', out_path)

# Parmetros
DIR_IMG = '/content/drive/MyDrive/Dados_Principios/MSI_original - (p)TOA'
DEST = '/content/drive/MyDrive/Dados_Principios/MSI_corrigido - (p)SUP'
os.makedirs(DEST, exist_ok=True)

# Supondo que img_id j est definido:
ID_BASE = '20250410T131301_20250410T131354_T23KMQ_'

for banda in atm_parametros:
    img = f'{DIR_IMG}/{ID_BASE}{banda}.tif'
    ref_toa = abrir_imagem(img)

    #QUANTIFICATION_VALUE = 10000
    #RADIO_ADD_OFFSET = -1000
    #ref_toa = (arr + RADIO_ADD_OFFSET) / QUANTIFICATION_VALUE

    p = atm_parametros[banda]
    Tatm = atm_parametros[banda]['T_upward'] * atm_parametros[banda]['T_downward']
    Tg = atm_parametros[banda]['tg_OG_co'] * atm_parametros[banda]['tg_OG_c02'] * atm_parametros[banda]['tg_OG_o2'] * atm_parametros[banda]['tg_OG_no2'] * atm_parametros[banda]['tg_OG_ch4']
    Tg_O3 = atm_parametros[banda]['Tg_O3']
    Tg_H20 = atm_parametros[banda]['Tg_H20']
    p_atm = atm_parametros[banda]['p_atm']
    s = atm_parametros[banda]['s']

    A = ((ref_toa / (Tg * Tg_O3)) - p_atm) / (Tatm * Tg_H20)
    ref_sup = A / (1 - (A * s))

    salvar_imagem(ref_sup, img, DEST)
```

## Comparao entre NDVI's e figuras

```python
# Variveis
DIR_IMG = '/content/drive/MyDrive/Dados_Principios/MSI_original - (p)TOA'
DEST = '/content/drive/MyDrive/Dados_Principios/MSI_corrigido - (p)SUP'

N = f'{DIR_IMG}/20250410T131301_20250410T131354_T23KMQ_B8.tif'
R = f'{DIR_IMG}/20250410T131301_20250410T131354_T23KMQ_B4.tif'

N_c = f'{DEST}/20250410T131301_20250410T131354_T23KMQ_B8_corr.tif'
R_c = f'{DEST}/20250410T131301_20250410T131354_T23KMQ_B4_corr.tif'

# Funo para calcular o NDVI
def calculate_ndvi(nir_band_path, red_band_path):
    with rasterio.open(nir_band_path) as nir_dataset, \
         rasterio.open(red_band_path) as red_dataset:

        # Read the NIR and Red bands as numpy arrays
        nir = nir_dataset.read(1).astype(float)
        red = red_dataset.read(1).astype(float)

        # Avoid division by zero
        ndvi = np.where(
            (nir + red) == 0.,
            0,
            (nir - red) / (nir + red)
        )

        # Get the metadata and transform
        profile = nir_dataset.profile

        # Update the data type to float32 for NDVI
        profile.update(
            dtype=rasterio.float32,
            count=1,
            compress='lzw')

        return ndvi, profile

# Calcular o NDVI para as bandas originais
ndvi_original, profile_original = calculate_ndvi(N, R)
output_original_ndvi = '/content/drive/MyDrive/Dados_Principios/ndvi_original.tif'
with rasterio.open(output_original_ndvi, 'w', **profile_original) as dst:
    dst.write(ndvi_original, 1)

print(f"NDVI original calculado e salvo em: {output_original_ndvi}")

# Calcular o NDVI para as bandas corrigidas
ndvi_corrected, profile_corrected = calculate_ndvi(N_c, R_c)
output_corrected_ndvi = '/content/drive/MyDrive/Dados_Principios/ndvi_corrected.tif'
with rasterio.open(output_corrected_ndvi, 'w', **profile_corrected) as dst:
    dst.write(ndvi_corrected, 1)

print(f"NDVI corrigido calculado e salvo em: {output_corrected_ndvi}")
```
